EJERCIO GRÁFICA SENO Y COSENO

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 6 * np.pi, 300)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)  # Primera subgráfica para el seno
plt.plot(x, y, "--b", ms=4, mec="r")
plt.title("Seno")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(color="green", linestyle="--", linewidth=0.4)

plt.subplot(1, 2, 2)  # Segunda subgráfica para el coseno
plt.plot(x, z, "--g", ms=4, mec="r")
plt.title("Coseno")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(color="green", linestyle="--", linewidth=0.4)

plt.suptitle("Gráficas de Seno y Coseno")
plt.tight_layout()
plt.show()

###FIN DEL PROGRAMA
```
