```python
def convert_length():
    print("Opciones de conversión de longitud:")
    print("1. SI a Sistema Inglés")
    print("2. Sistema Inglés a SI")
    option = int(input("Selecciona una opción: "))

    try:
        if option == 1:
            length_si = float(input("Ingrese la longitud en metros o kilómetros: "))
            target_unit = input("Ingrese la unidad (pies, millas, yardas): ").lower()
            if target_unit == "pies":
                length_english = length_si * 3.28084
                print(f"Longitud en Sistema Inglés: {length_english} pies")
            elif target_unit == "millas":
                length_english = length_si / 1.60934
                print(f"Longitud en Sistema Inglés: {length_english} millas")
            elif target_unit == "yardas":
                length_english = length_si * 1.09361
                print(f"Longitud en Sistema Inglés: {length_english} yardas")
            else:
                print("Unidad no válida.")
        elif option == 2:
            length_english = float(input("Ingrese la longitud en pies, millas o yardas: "))
            source_unit = input("Ingrese la unidad (pies, millas, yardas): ").lower()
            if source_unit == "pies":
                length_si = length_english / 3.28084
                print(f"Longitud en SI: {length_si} metros")
            elif source_unit == "millas":
                length_si = length_english * 1.60934
                print(f"Longitud en SI: {length_si} kilómetros")
            elif source_unit == "yardas":
                length_si = length_english / 1.09361
                print(f"Longitud en SI: {length_si} metros")
            else:
                print("Unidad no válida.")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Ingrese un valor numérico válido.")

def convert_mass():
    print("Opciones de conversión de masa:")
    print("1. SI a Sistema Inglés")
    print("2. Sistema Inglés a SI")
    option = int(input("Seleccione una opción: "))

    try:
        if option == 1:
            mass_si = float(input("Ingrese la masa en kilogramos o gramos: "))
            target_unit = input("Ingrese la unidad de destino (libras, onzas): ").lower()
            if target_unit == "libras":
                mass_english = mass_si * 2.20462
                print(f"Masa en Sistema Inglés: {mass_english} libras")
            elif target_unit == "onzas":
                mass_english = mass_si * 35.274
                print(f"Masa en Sistema Inglés: {mass_english} onzas")
            else:
                print("Unidad no válida.")
        elif option == 2:
            mass_english = float(input("Ingrese la masa en libras o onzas: "))
            source_unit = input("Ingrese la unidad (libras, onzas): ").lower()
            if source_unit == "libras":
                mass_si = mass_english / 2.20462
                print(f"Masa en SI: {mass_si} kilogramos")
            elif source_unit == "onzas":
                mass_si = mass_english / 35.274
                print(f"Masa en SI: {mass_si} gramos")
            else:
                print("Unidad no válida.")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Ingrese un valor numérico válido.")

def convert_time():
    print("Opciones de conversión de tiempo:")
    print("1. Unidades más grandes a unidades más pequeñas")
    print("2. Unidades más pequeñas a unidades más grandes")
    option = int(input("Seleccione una opción: "))

    try:
        if option == 1:
            time_si = float(input("Ingrese la cantidad de tiempo: "))
            target_unit = input("Ingrese la unidad de destino (segundos, minutos, milisegundos): ").lower()
            if target_unit == "segundos":
                time_english = time_si * 3600
                print(f"Tiempo en Sistema Inglés: {time_english} segundos")
            elif target_unit == "minutos":
                time_english = time_si * 60
                print(f"Tiempo en Sistema Inglés: {time_english} minutos")
            elif target_unit == "milisegundos":
                time_english = time_si * 3600000
                print(f"Tiempo en Sistema Inglés: {time_english} milisegundos")
            else:
                print("Unidad de destino no válida.")
        elif option == 2:
            time_english = float(input("Ingrese la cantidad de tiempo en segundos, minutos o milisegundos: "))
            source_unit = input("Ingrese la unidad de origen (segundos, minutos, milisegundos): ").lower()
            if source_unit == "segundos":
                time_si = time_english / 3600
                print(f"Tiempo en SI: {time_si} horas")
            elif source_unit == "minutos":
                time_si = time_english / 60
                print(f"Tiempo en SI: {time_si} minutos")
            elif source_unit == "milisegundos":
                time_si = time_english / 3600000
                print(f"Tiempo en SI: {time_si} segundos")
            else:
                print("Unidad de origen no válida.")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Ingrese un valor numérico válido.")

def convert_volume():
    print("Opciones de conversión de volumen:")
    print("1. Galones a litros")
    print("2. Litros a galones")
    option = int(input("Selecciona una opción: "))

    try:
        if option == 1:
            gallons = float(input("Ingrese la cantidad en galones: "))
            liters = gallons * 3.78541
            print(f"Volumen en litros: {liters} litros")
        elif option == 2:
            liters = float(input("Ingrese la cantidad en litros: "))
            gallons = liters / 3.78541
            print(f"Volumen en galones: {gallons} galones")
        else:
            print("Opción no válida.")
    except ValueError:
        print("Ingrese un valor numérico válido.")

while True:
    print("Menú de conversiones:")
    print("1. Convertir unidades de longitud")
    print("2. Convertir unidades de masa")
    print("3. Convertir unidades de tiempo")
    print("4. Convertir unidades de volumen")
    print("5. Salir del programa")

    choice = input("Seleccione una opción: ")

    if choice == '1':
        convert_length()
    elif choice == '2':
        convert_mass()
    elif choice == '3':
        convert_time()
    elif choice == '4':
        convert_volume()
    elif choice == '5':
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida.")

#FiN DEL PROGRAMA
```
