**1 - Numero Primo**

```python
def el_numero_es__primo(numero):

    # Si el número es menor o igual a 1, manda un mensaje de error
    if numero <= 1:
        return "Error en el Número"

    # Repite desde 2 hasta la raíz cuadrada del número + 1
    for i in range(2, int(numero**0.5) + 1):
        # Si el número es divisible por algún otro número, no es primo
        if numero % i == 0:
            return False

    # Si no se encontraron divisores entonces el número es primo
    return True

def main():
    while True:
        try:
            # Solicitar ngresar un número entero
            num = int(input("Ingrese un número entero mayor que 1: "))

            # Llama a la función es_primo y almacena el resultado
            resultado = el_numero_es__primo(num)

            # Imprime el resultado
            print(resultado)

            # Si el resultado no es un error, termina el bucle
            if resultado != "Error en el Número":
                break

        except ValueError:
            # Mensaje de error si el usuario ingresa algo que no es un número entero
            print("Ingrese un número entero válido.")

if __name__ == "__main__":
    # Llama a la función principal
    main()

### FIN DEL PROGRAMMA
```

**2 - Arreglo de números**

```python
import matplotlib.pyplot as plt

def calcular_promedio(datos):
    return sum(datos) / len(datos)

def calcular_varianza(datos, promedio):
    return sum((x - promedio) ** 2 for x in datos) / len(datos)

def calcular_desviacion_estandar(varianza):
    return varianza ** 0.5

# Crear un arreglo de números enteros entre 0 y 150
numeros = [i for i in range(151)]

# Calcular el promedio
promedio = calcular_promedio(numeros)

# Calcular la varianza
varianza = calcular_varianza(numeros, promedio)

# Calcular la desviación estándar
desviacion_estandar = calcular_desviacion_estandar(varianza)

# Ver los datos en un histograma
plt.hist(numeros, bins=20, color='magenta', edgecolor='yellow')
plt.title('Histograma de Números')
plt.xlabel('Número')
plt.ylabel('Frecuencia')
plt.savefig('Histograma de Números.png',dpi=300,format='png')
plt.show()

# Calcular la suma de números pares e impares
suma_pares = sum(x for x in numeros if x % 2 == 0)
suma_impares = sum(x for x in numeros if x % 2 != 0)

# Calcular los múltiplos de 2, 3 y 5
multiplos_dos = list(filter(lambda x: x % 2 == 0, numeros))
multiplos_tres = list(filter(lambda x: x % 3 == 0, numeros))
multiplos_cinco = list(filter(lambda x: x % 5 == 0, numeros))

print(f'Promedio: {promedio}')
print(f'Varianza: {varianza}')
print(f'Desviación Estándar: {desviacion_estandar}')
print(f'Suma de números pares: {suma_pares}')
print(f'Suma de números impares: {suma_impares}')
print(f'Múltiplos de 2: {multiplos_dos}')
print(f'Múltiplos de 3: {multiplos_tres}')
print(f'Múltiplos de 5: {multiplos_cinco}')

### FIN DEL PROGRAMA
```

**3 - Familia de curvas para la solución de la ecuación diferencial**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función dee la solución general
def solucion_general(x, C):
    return (x - 1) + C * np.exp(-x)

# Valores de x
x_values = np.linspace(-2, 5, 400)

# Graficar para valores de C>0
for C_positivo in np.linspace(0.1, 2, 5):
    y_values = solucion_general(x_values, C_positivo)
    plt.plot(x_values, y_values, label=f'C > 0, C={C_positivo}', color='magenta')

# Graficar para C=0
y_values_cero = solucion_general(x_values, 0)
plt.plot(x_values, y_values_cero, label='C = 0', color='green')

# Graficar para valores de C<0
for C_negativo in np.linspace(-2, -0.1, 5):
    y_values = solucion_general(x_values, C_negativo)
    plt.plot(x_values, y_values, label=f'C < 0, C={C_negativo}', color='yellow')

# Etiquetas
plt.title('Familia de Curvas para la Solución de la Ecuación Diferencial')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('Gráfica Familia de Curvas.png',dpi=300,format='png')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

**4 - Funcion**


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return 3*x + 7*x**2 - 4

# Valores de x
x_values = np.linspace(-10, 10, 1000)

# Valores de la función
y_values = f(x_values)

# Graficar la función
plt.plot(x_values, y_values, label='3x + 7x^2 - 4', color='blue')

# Graficar las asíntotas horizontales y verticales
plt.axhline(0, color='black', linestyle='--', label='Asíntota horizontal')
plt.axvline(0, color='red', linestyle='--', label='Asíntota vertical')

# Establecer límites en el eje y para evitar valores muy altos o bajos
plt.ylim(-100, 100)

plt.legend()

# Etiquetas y título
plt.title('Gráfica de la función con asíntotas')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)
plt.show()
```
